```python
# Linear Graph 
import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot Distance and Time

distance = list(range(2,11,2))
time = list(range(1,6))
plt.title('Distance Vs Time')
d = 2 * time
plt.plot(time,distance)
plt.show()
```

```python
distance = list(range(2,11,2))
time = list(range(1,6))
plt.title('Distance Vs Time')
d = 2 * time
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# Non Linear graph
# Velocity = acceleration * time
t = list(range(1,11))

a = list(range(len(t)+1))

v = [a[i]*i for i in t]
plt.title('Velocity Vs Time')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.grid()
plt.plot(t,v)
plt.show()
```

```python
n = list(range(10))
x = [i**3 for i in n]
plt.plot(n,x)
plt.grid()
```

```python
# Voltege vs Time
import numpy as np
#t = [i/10 for i in range(1,21)]
t = np.arange(0.1,2.0,0.1)
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.title('Time Vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
years = df.columns[2:]
incomeArkansas = df.values[3,2:]
plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Plot Average Income of all States from 2005 to 2013

df
incomeallstates = []
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())

incomeallstates
plt.title('Income of All States')
plt.plot(years,incomeallstates)
plt.grid()
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```
